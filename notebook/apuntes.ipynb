```python

import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
import numpy as np
```

```python
#Grafico de distribucion normal

# Tu array de distribución normal
array_normal = np.random.normal(20, 3, 1000)  # Ejemplo con media 0 y desviación estándar 1


# Punto al que deseas centrar la vista
x_centro = 20  # Por ejemplo, centrar en x=3
y_centro = 0  # Por ejemplo, centrar en y=10

# Establecer límites del eje x y y para centrar la vista en (x_centro, y_centro)
mitad_ancho_vista_x = 10  # La mitad del ancho de la vista que deseas en el eje x
mitad_alto_vista_y = 15  # La mitad de la altura de la vista que deseas en el eje y

plt.xlim(x_centro - mitad_ancho_vista_x, x_centro + mitad_ancho_vista_x)
plt.ylim(y_centro, y_centro + mitad_alto_vista_y)


# Graficar un histograma
plt.hist(array_normal, bins=200, color='skyblue', edgecolor='black')
plt.title('Distribución Normal')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

```python

# Crea el slider
slider = widgets.FloatSlider(
    value=1.0,
    min=0,
    max=100,
    step=1,
    description='Valor del Slider:'
)


# Función para actualizar el gráfico con el valor del slider
def actualizar_grafico(valor_slider):
    # Borra el gráfico anterior si existe
    plt.clf()
    
    #INTRODUCES EL GRAFICO QUE QUIERAS ALTERAR DENTRO DE LA FUNCION

    df = valor_slider
    chi_squared_values = np.random.chisquare(df, size=1000)

    # Graficar un histograma
    plt.hist(chi_squared_values, bins=200, color='skyblue', edgecolor='black')
    plt.title('Distribución Normal')
    plt.xlabel('Valores')
    plt.ylabel('Frecuencia')
    plt.grid(True)

    # Mostrar el gráfico
    plt.show()


# Conecta la función de actualización al slider utilizando widgets.interactive
interactive_plot = widgets.interactive(actualizar_grafico,
                                        valor_slider=slider)

# Muestra el slider y el gráfico inicialmente
interactive_plot
```
